```python
ax = sns.violinplot(y='mpg', x='origin', hue='year', data=df)
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
```

```python
ax = sns.boxenplot(y='mpg', x='origin', hue='year', data=df)
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
```

```python
ax = sns.boxenplot(y='weight', x='origin', hue='year', data=df)
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
```

```python
df['mpg_per_weight'] = df['mpg'] / df['weight']
```

```python
ax = sns.boxenplot(y='mpg_per_weight', x='origin', hue='year', data=df)
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
```
